```python
import array

arr = array.array('i', [1, 2, 3, 4, 5])
a, *rest, e = arr
print(a)     # 1
print(rest)  # [2, 3, 4]
print(e)     # 5
```

```python
import array

arr1 = array.array('i', [1, 2])
arr2 = array.array('i', [3, 4])
arr3 = [arr1, arr2]

(a, b), (c, d) = arr3
print(a, b)  # 1 2
print(c, d)  # 3 4
```

```python
import array

arr = array.array('i', [10, 20, 30, 40, 50])
first_two, *rest = arr[:2], arr[2:]
print(first_two)  # [10, 20]
print(rest)       # [30, 40, 50]
```
